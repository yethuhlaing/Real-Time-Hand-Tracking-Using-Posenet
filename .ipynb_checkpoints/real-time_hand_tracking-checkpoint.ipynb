## Importing dependencies

```python
import cv2 
import mediapipe as mp
import time
```

## Making Camera Object

```python
cap = cv2.VideoCapture(0)
while True:
    success , image = cap.read()
    cv2.imshow("Hand Tracking", image)
    
    if cv2.waitKey(10) & 0xFF == ord("q"):
        break
    
cap.release()
cv2.destroyAllWindows()
```

# Create Object from Class Hand

```python
cap = cv2.VideoCapture(0)

# Hand Object
# It has four default parameters
# static_image_mode => true - tracks and detects => false - sometimes tracks and detects
# max_num_hands
# min_detection_confidence
# min_tracking_confidence
hands = mp.solutions.hands.Hands()

while True:
    success , image = cap.read()
    
    # Converting RGB
    img_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # Getting the hand image
    results = hands.process(img_RGB)
    
    # Showing the Image
    cv2.imshow("Hand Tracking", image)
    
    if cv2.waitKey(10) & 0xFF == ord("q"):
        break
    
cap.release()
cv2.destroyAllWindows()
```

# Check if the Hand is being detected or not

```python
cap = cv2.VideoCapture(0)

# Hand Object
hands = mp.solutions.hands.Hands()

while True:
    success , image = cap.read()
    
    # Converting RGB
    img_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # Getting the hand image
    results = hands.process(img_RGB)
    print(results.multi_hand_landmarks)
    # Showing the Image
    cv2.imshow("Hand Tracking", image)
    
    if cv2.waitKey(10) & 0xFF == ord("q"):
        break
    
cap.release()
cv2.destroyAllWindows()
```

# Detecting Landmarks and Draw points on Hand

```python
cap = cv2.VideoCapture(0)
hands = mp.solutions.hands.Hands()
mp_draw = mp.solutions.drawing_utils

while True:
    success , image = cap.read()
    img_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = hands.process(img_RGB)
    
    # if landmarks are deteted, lines are drawn
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_draw.draw_landmarks(image, hand_landmarks)
    
    cv2.imshow("Hand Tracking", image)
    if cv2.waitKey(10) & 0xFF == ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```

# Draw Connection between Landmarks

```python
cap = cv2.VideoCapture(0)

mp_hands = mp.solutions.hands
mp_draw = mp.solutions.drawing_utils

hands = mp_hands.Hands()

while True:
    success , image = cap.read()
    img_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = hands.process(img_RGB)
    
    # Add "HAND_CONNECTIONS" parameter
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_draw.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
    
    cv2.imshow("Hand Tracking", image)
    if cv2.waitKey(10) & 0xFF == ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```

# Extracting value of each landmark

```python
cap = cv2.VideoCapture(0)

mp_hands = mp.solutions.hands
mp_draw = mp.solutions.drawing_utils
hands = mp_hands.Hands()

c_time = 0
p_time = 0
while True:
    success , image = cap.read()
    img_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = hands.process(img_RGB)
    
    if results.multi_hand_landmarks:
        # Getting circle values 
        for hand_landmarks in results.multi_hand_landmarks:
            for id , lm in enumerate(hand_landmarks.landmark):
                h ,w , c = image.shape
                cx , cy = int(lm.x*w), int(lm.y*h)
                print(id, cx ,cy)
            mp_draw.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
        
    #display FPS
    c_time = time.time()
    fps = 1/(c_time-p_time)
    p_time = c_time
    cv2.putText(image, str(int(fps)), (10, 70), cv2.FONT_HERSHEY_COMPLEX, 3 , (255, 0, 255), 3)
    
    cv2.imshow("Hand Tracking", image)
    if cv2.waitKey(10) & 0xFF == ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```

# Hightling Fingertips

```python
cap = cv2.VideoCapture(0)

mp_hands = mp.solutions.hands
mp_draw = mp.solutions.drawing_utils
hands = mp_hands.Hands()

c_time = 0
p_time = 0
while True:
    success , image = cap.read()
    img_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = hands.process(img_RGB)
    
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            for id , lm in enumerate(hand_landmarks.landmark):
                h ,w , c = image.shape
                cx , cy = int(lm.x*w), int(lm.y*h)
                
                # Drawing Circle
                if id == 4:
                    cv2.circle(image, (cx ,cy) , 15, ( 255, 255 ,0) , cv2.FILLED)
                if id == 8:
                    cv2.circle(image, (cx ,cy) , 15, ( 255, 255 ,0) , cv2.FILLED)
                if id == 12:
                    cv2.circle(image, (cx ,cy) , 15, ( 255, 255 ,0) , cv2.FILLED)
                if id == 16:
                    cv2.circle(image, (cx ,cy) , 15, ( 255, 255 ,0) , cv2.FILLED)
                if id == 20:
                    cv2.circle(image, (cx ,cy) , 15, ( 255, 255 ,0) , cv2.FILLED)
            mp_draw.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
        
    #display FPS
    c_time = time.time()
    fps = 1/(c_time-p_time)
    p_time = c_time
    cv2.putText(image, str(int(fps)), (10, 70), cv2.FONT_HERSHEY_COMPLEX, 3 , (255, 0, 255), 3)
    
    cv2.imshow("Hand Tracking", image)
    if cv2.waitKey(10) & 0xFF == ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```
